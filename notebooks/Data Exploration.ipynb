<h1><center>Exploratory Data Analysis of Wine Data</center></h1>

This datasets is related to red variants of the Portuguese "Vinho Verde" wine. The dataset describes the amount of various chemicals present in wine and their effect on it's quality. The purpose of this notebook report is to take a deep dive analysis of the dataset prior to modelling.

It is an important precautionary step to understand the dataset in hand before training it in a model, as by analyzing we may come across some interesting findings that can guide us in future during modelling stage of this project.

## Import Libraries

```python
df = pd.read_csv("../data/clean_data/train.csv", usecols=range(1, 13))
```

## Independent Variable Analysis

```python
ind_var = df.columns[:-1]
```

Since all the independent variables are continuous, we can take a look at the distribution.

```python
with warnings.catch_warnings(record=True):
    fig, axes = plt.subplots(nrows=11, ncols=1, figsize=(5,50))
    plt.subplots_adjust(hspace=0.3)
    fig.suptitle("Histogram of independent variables", fontsize=18, y=0.90)

    for i, column in enumerate(ind_var):
        sns.distplot(df[column],ax=axes[i])
        axes[i].set_title(column)
        axes[i].set(xlabel=None)
```

Analyse the skewness of the variable, skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. Having an unskewed dataset can really improve the model's predicition especially in linear models.

```python
df[ind_var].skew(axis=0)
```

From the above two observations, we can notice that:
<ol>
  <li>fixed_acidity: follows normal distribution slightly with a small skew towards right.</li>
  <li>volatile_acidity: follows normal distribution somewhat closely with relatively small skew.</li>
  <li>citric_acid: does not follow a normal distribution.</li>
  <li>residual_sugar: follows normal distribution slightly with a heavy skew towards right.</li>
  <li>chlorides: follows normal distribution slightly with a heavy skew towards right.</li>
  <li>free_sulfur_dioxide: does not follow a normal distribution with a small skew towards right.</li>
  <li>total_sulfur_dioxide: follows normal distribution very slightly with a skew towards right.</li>
  <li>density: follows normal distribution somewhat closely with relatively small skew.</li>
  <li>pH: follows normal distribution somewhat closely with relatively small skew.</li>
  <li>sulphates: follows normal distribution somewhat closely with a skew towards right.</li>
  <li>alcohol: does not follow a normal distribution with a small skew towards right.</li>
</ol>

We can conduct pairwise comparison of each independent variable to find out how each work with other variables. By finding out high positive correlated variables, we can avoid one of them to avoid overfitting as both variables provide more or less same information. 

```python
df[ind_var].corr()
```

From the above, fixed_acidity and citric_acid has the greatest pairwise correlation (0.70) followed by total_sulfur_dioxide and free_sulfur_dioxide (0.64).

## Dependent Variable Analysis

It is also important, to take a look at the target variable, as if there is an imbalance in our training data it can severely affect the model's robustness.

```python
fig, ax = plt.subplots(figsize = (25,10))
sns.countplot(x= df["quality"])
plt.title("Wine Quality Count")

total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
```

From the above count plot, we can see an imbalance in our training data. Wine Quality with 5 and 6 account for 40.7% and 38.0%, respectively, from the train data. Whereas quality 3, 4, and 8 have significantly fewer data.

## Relationship between Independent and Dependent Variable

Independent variables that has a high correlation with target variable is a good predictor variable, and should be considered during model fitting.

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df.corr()[["quality"]].sort_values('quality').tail(14), vmax=1, vmin=-1, cmap='YlGnBu', annot=True, ax=ax)
ax.invert_yaxis()
```

From the above heat map, we can see that:
* Significant Positive Correlation
    1. alcohol (0.48)
    2. sulphates (0.4)
    3. citric_acid (0.14)
* Significant Negative Correlation
    1. volatile_acidity (-0.21)
    2. total_sulfur_dioxide (-0.22)
    3. density (-0.15)

## Multicollinearity

```python
vif_data = pd.DataFrame()
vif_data["feature"] = df[ind_var].columns

vif_data["VIF"] = [variance_inflation_factor(df[ind_var].values, i)
                          for i in range(len(df[ind_var].columns))]
  
print(vif_data)
```

VIF values for all the indpendent variables are all very high, we should either drop or combine variables.

## Key Takeaways

We should consider the findings from this notebook during feature engineering and model building part of the project.
### Feature Engineering:
* **Feature Selection**: We should consider independent variables that has a high correlation with dependent variable, if the variable is highly correlated to any other variable ignore the variable that has a lower correlation to dependent variable. VIF score is something that needs to be cosidered too.
    1. Following columns have to be dropped: fixed_acidity, free_sulfur_dioxide, pH, chlorides, residual_sugar
    2. Perform Principal Component Analysis on remaining features that have a high VIF score
* **Distribution and Skewness**: We should consider performing transformations on variables that do not follow a normal distribution. In addition, since independent variables with heavy skewness, we should also consider performing transformations on those variables.
    1. Perform transformations on total_sulfur_dioxide, alcohol, citric_acid
    
### Model Building:
* **Model Choice**: choose models that are more robust in dealing with imabalance of classes and multicollinearity.
* **Oversampling Minority Class**: data of wine quality 3, 4, 7, and 8 are less and have to make synthetic data to accomodate imbalance 
* **Evaluation Metric**: Precision and F1 Score should be considered instead of accuracy.
